```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
crime_data = "crime_data.csv"

# Read the crime_data
crime_df = pd.read_csv(crime_data)
crime_df.head()
```

```python
# drop rows that don't have a value in the "OCCURRED ON" column
crime_df.dropna(subset=['OCCURRED ON'])
# preview dataframe with edited date
crime_df.head()
```

```python
# group the data by the day in which the crime occurred
crime_by_date = crime_df.groupby(['OCCURRED ON'])
len(crime_by_date)
crime_by_date.head()
```
