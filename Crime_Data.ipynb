```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
crime_data = "crime_data.csv"

# Read the crime_data
crime_df = pd.read_csv(crime_data)
crime_df.head()
```

```python
# create new data frame with the columns that will be helpful for the project: 
#inc number, occurred on, UCR crime cateogry, and premise type
new_crime_df = pd.DataFrame({'INC NUMBER': crime_df['INC NUMBER'],
                'OCCURRED ON': crime_df['OCCURRED ON'],
                'UCR CRIME CATEGORY': crime_df['UCR CRIME CATEGORY'],
                'PREMISE TYPE': crime_df['PREMISE TYPE']})
new_crime_df.head()
```

```python
# Take out the time from the "OCCURRED ON" column so you can group data by day
new_crime_df['OCCURRED ON'] = new_crime_df['OCCURRED ON'].map(lambda x: str(x)[:-6])
new_crime_df['OCCURRED ON']
```

```python
# drop rows that don't have a value
new_crime_df.dropna()
```
